## 메모리의 이해

````python
# id(변수)는 메모리 '주소'가 나오므로, 아래 예시와는 다른 값이 나올 것이다.
a = 10
b = a
print(id(a))
print(id(b))
b = b + 1
print(id(a))
print(id(b))
b = b - 1
print(id(b))
a = [1,2,3]
b = 3
print(id(a))
print(id(a[2]))
print(id(b))
print(id([1,2,3]))
print(b == a[2])
print(b is a[2])
print(a == [1,2,3])
print(a is not [1,2,3])
b = a
print(id(a))
print(id(b))
b[0] = 4
print(a)
print(b)
````

### 마린 
```python
<marine.py>
class Marine:
    attack_damage = 10
    def __init__(self):
        self.hp = 100
        self.attack_rate = 5
    def fire(self, target):
        target.hp -= Marine.attack_damage
    def stimpack(self):
        self.hp -= 15
        self.attack_rate += 20
    def sayhello():
        print('hello')
        
def sayhello():
    print("global hello")

Marine.sayhello()
m1 = Marine()
m1.fire(m1)
print(m1.hp)
```
