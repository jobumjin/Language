## 고급 자료형, 고급 기법, 재귀 함수
* 자료형 : 딕셔너리

```python
#딕셔너리 생성
dict1 = {"key1":"value1","key2":"value2"}
print(dict1)
print(dict1["key1"])
#print(dict1["key3"]) Error
dict1 = {1:"value1",2:"value2"}
print(dict1[2])
dict1 = {1:1,"key2":"value2"}
print(dict1[1])
print(dict1["key2"])
dict1 = {1:[1,2,3],2:[4,5,6]}
print(dict1[2])
print(dict1[2][1])
dict1 = dict()
#print(dict1[0]) Error
dict1 = {1:1 ,1:2, 1:3}
print(dict1)
print(dict1[1])
#dict1 = {(1,2):"hello"} Error
dict1 = {"k1":{"k2":"value"}}
print(dict1)
print(dict1["k1"]["k2"])
```

```python
#딕셔너리 item 추가/삭제
dict1 = {1:"value1"}
print(dict1)
dict1[2] = "value2"
print(dict1)
dict1[1] = 123
print(dict1)
dict1[(1,2,3)] = "value3"
print(dict1)
del dict1[1]
print(dict1)
#del dict1[1] Error
del dict1[(1,2,3)]
print(dict1)
```

```python
#딕셔너리 관련 기타 함수들
dict1 = {1:123,"2":"456"}
print(dict1.keys())
for k in dict1.keys():
    print(k)
print(list(dict1.keys()))
print(dict1.values())
print(dict1.items())
print(dict1.get(1))
print(dict1.get(2))
print(dict1.clear())
```

## 자료형 셋(집합)

```python
#셋 생성
s = set([1,2,3])
print(s)
#print(s[0]) Error
s = set("hello")
print(s)
s = set(['h','e','l','l','o'])
print(s)
```

```python
#셋을 리스트/튜플로 변환
s = set("hello")
li = list(s)
tu = tuple(s)
print(s)
print(li)
print(tu)
print(tu[1])
```

```python
#합집합, 교집합, 차집합
s1 = {1,2,3,4,4,4}
s2 = set([3,4,5,6])
print(s1 | s2)
print(s1.union(s2))
print(s1 & s2)
print(s1.intersection(s2))
print(s1 - s2)
print(s2 - s1)
print(s1.difference(s2))
print(s2.difference(s1))
```

```python
s = set([1,2,3])
print(s)
s.add(4.5)
print(s)
s.add("hello")
print(s)
temp = set("hello")
#s.add(temp) Error
s.update([4,5,6])
print(s)
s.update([0,1,2])
print(s)
s.remove(0)
print(s)
#s.remove([1,2])Error
```

## 고급기법들

```python
#3항 연산
a = 3
result = 'odd' if a% 2 == 1 else 'even'
print(result)
```

```python
#for 문 다중 반복자
a = [(1,2),(3,4)]
for x in a:
    print(x)
for x1, x2 in a:
    print(x1 + x2)
```

```python
#무시(_)
def sum_mul(x1,x2):
    return(x1+x2, x1*x2)
s, m = sum_mul(2,3)
print(s,m)
r = sum_mul(7,9)
print(r)
a = [(1,2),(3,4)]
for x1, _ in a:
    print(x1*2)
```

```python
#List comprehension
m = [1,2,3,4]
subm1 = [r*2 for r in m]
print(subm1)
subm2 = [r*2 for r in m if r % 2 == 0]
print(subm2)
m = [[1,2],[3,4]]
subm1 = [r*2 for r in m]
print(subm1)
subm2 = [i for row in m for i in row]
print(subm2)
```

```python
#Dictionary comprehension
d = {"a":10,"b":34}
reverse_d = {v:k for k, v in d.items()}
print(reverse_d)
d2 = {k:k*2 for k in d.keys()}
print(d2)
```

```python
#Set comprehension
s = {3,1,4}
print(s)
s2 = {x**2 for x in s}
print(s2)
```

```python
#String format
a = 3
print("a = {}".format(a))
print("a = {}".format(float(a)))
a = 3; b = 5.0
print("a = {},b = {}".format(a,b))
print("b = {1},a = {0}".format(a,b))
print("a is %d" % a)
print("b is %d" % b)
print("a is %d, b is %f" % (a,b))
r1 = "a is %d, b is %f" % (a,b)
r2 = "a*b = {}".format(a*b)
print(r1,r2)
print("%0.4f"%3.141592)
print("%10.4f"%3.141592)
```

* %s 문자열 (String)
* %c 문자 1개 (Character)
* %d 정수 (Integer)
* %f 부동소수 (Floating-point)
* %o 8진수
* %x 16진수
* %%  Literal % (문자 % 자체)

```python
#Enumerate
a = ['a','b','c']
for i, x in enumerate(a):
    print("[{0}]{1}".format(i,x))
```

## 재귀함수

```python
#반복문과 재귀함수의 관계
def func():
    print("hello")
    func()
func()
```

```python
#재귀함수로 구구단 2단 출력
def func(n):
    if n > 9: return
    print(2 * n)
    func(n+1)
func(1)
```

```python
#재귀함수로 1~5 합 구하기
def getsum(n):
    if n < 1:return 0
    return n + getsum(n-1)
print(getsum(5))
```

```python
#연습
def func(d,k):
    if not d.get(k):
        return k
    return func(d,d[k])
print(func({"a":'b','b':'c',"c":"d"},"a"))
```
